```python
import numpy as np
import pandas as pd
df = pd.read_csv('worldfloras.csv').dropna()
df
```

```python
pattern=re.compile(r'[DE][^a].*')
matches=pattern.finditer(df["Country"].to_string())
for match in matches:
  print(match)
```

```python
pattern=re.compile(r'\bNew.*')
matches=pattern.finditer(df["Country"].to_string())
for match in matches:
  print(match)
```

```python
pattern=re.compile(r'[A-Z][y].*')
matches=pattern.finditer(df["Country"].to_string())
for match in matches:
  print(match)
```

```python
def print_words(list_words,numElementsRow):
  for j in range (0,len(list_words),numElementsRow):
    for i in range(j,j+numElementsRow):
      if (i>=len(list_words)):
        break
      print(i,'.',list_words[i],end=',')
    print()
  print()
```
