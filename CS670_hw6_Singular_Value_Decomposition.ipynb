### Verify that by multiplying (Matrix multiplication) of the 3 SVD matrices will result in the original matrix

```python
SigmaMatrix = np.diag(Sigma)
print(" Sigma Diagonal Matrix:\n",SigmaMatrix)
```

```python
recreateA =np.dot(np.dot(V,SigmaMatrix),Ustar)
print("Recreated Matrix:\n",recreateA)
```

# Two rows are linear derivation of some other rows. So the Sixma matrix shows that dependency

image.png

```python
A1=np.zeros(shape=(2,3))
A2=np.zeros(shape=(3,2))
A1[0] = [3,1,1]
A1[1] = [-1,3,1]

A2[0] = [1,1]
A2[1] = [-0,1]
A2[2] = [-1,1]
```

```python
V1, Sigma1, Ustar1=svd(A1)
V2, Sigma2, Ustar2=svd(A2)
```

```python
print("V1:\n",V1,"\n Sigma1:\n",Sigma1,"\n Ustar1\n",Ustar1)
```

```python
Sigma1d = np.diag(Sigma1)
z1=np.vstack((np.dot(V1,Sigma1d),np.zeros(shape=(1,2))))
z2=np.hstack((z1,np.zeros(shape=(3,1))))
recreateA1 =np.dot(z2,Ustar1)
print("Recreated Matrix:\n",recreateA1 )
print("Original Matrix:\n",A1 )
```

```python
print("V2:\n",V2,"\n Sigma2:\n",Sigma2,"\n Ustar2\n",Ustar2)
```

```python
Sigma2d = np.diag(Sigma2)
z1=np.vstack((Sigma2d,np.zeros(shape=(1,2))))
z2= np.dot(V2,z1)
recreateA2 =np.dot(z2,Ustar2)
print("Recreated Matrix:\n",recreateA2 )
print("Original Matrix:\n",A2 )
```
