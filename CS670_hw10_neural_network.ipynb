### Solution

```python
import tensorflow as tf
import numpy as np
import math
import pandas as pd
```

```python
x1=0.1
x2=0.2
w1=0.15
w2=0.05
w3=0.36
b1=0.33
b2=0.56

def activation(fn,type):
  df= pd.DataFrame()
  z1= x1*w1 + x2*w2 + b1
  if type=='kera' :
    a1=round(eval('tf.' + fn +'(z1)'),3)
  else:
    a1=round(eval('tf.' + fn +'(z1).numpy()'),3)
  df['a1']= pd.DataFrame([a1],index = [fn])
  z2=a1*w3 + b2
  if type=='kera' :
    a2=round(eval('tf.' + fn +'(z2)'),3)
  else:
    a2=round(eval('tf.' + fn +'(z2).numpy()'),3)
  df['a2']= pd.DataFrame([a2],index = [fn])
  return df
```

```python
df2 =pd.DataFrame()
df2=df2.append(activation('math.sigmoid','t'))
df2=df2.append(activation('keras.activations.linear','kera'))
df2=df2.append(activation('math.tanh','t'))
df2=df2.append(activation('nn.relu','t'))
df2
```

### Activation Function and Gradient Calculator

```python
x= ['-4.0',' 0.5',' 4.0']

def activation(fn):
  for i in x:
    t1=round(eval('tf.' + fn +'('+i+').numpy()'),3)
    if fn=='math.sigmoid':
      t2= round(t1* (1-t1),6)
    elif fn=='math.tanh':
      t2= round(1-pow(t1,2),6)
    else:
      if t1>0 :
        t2= 1
      else:
        t2=0
    print('function:',fn, 'x=',i,'gradient=', t2) 
```

```python
activation('math.tanh')
```

```python
activation('nn.relu')
```
